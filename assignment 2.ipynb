```python
#Q3. Solution of Q3 
#example of ..Mutable Objects
#List
my_list = [1, 2, 3]
my_list[0] = 10
print(my_list)  # Outputs: [10, 2, 3]

#Dictionaries
my_dict = {'a': 1, 'b': 2}
my_dict['a'] = 10
print(my_dict)  # Outputs: {'a': 10, 'b': 2}

#sets
my_set = {1, 2, 3}
my_set.add(4)
print(my_set)  # Outputs: {1, 2, 3, 4}

## examples of Immutable Objects
#Strings
my_string = "hello"
new_string = my_string.replace('h', 'y')
print(my_string)  # Outputs: "hello"
print(new_string)  # Outputs: "yello"

#Tuples
my_tuple = (1, 2, 3)
new_tuple = my_tuple + (4,)
print(my_tuple)  # Outputs: (1, 2, 3)
print(new_tuple)  # Outputs: (1, 2, 3, 4)

##Integers
my_int = 5
new_int = my_int + 3  # Adding 3 to my_int
print(new_int)  # Outputs: 8
```

```python
#Q4 Example with out output


#Arithmetic Operators
# Addition
a = 10
b = 5
print("Addition:", a + b)  # Output: 15

# Subtraction
print("Subtraction:", a - b)  # Output: 5

# Multiplication
print("Multiplication:", a * b)  # Output: 50

# Division
print("Division:", a / b)  # Output: 2.0

# Modulus
print("Modulus:", a % b)  # Output: 0

# Exponentiation
print("Exponentiation:", a ** b)  # Output: 100000

# Floor Division
print("Floor Division:", a // b)  # Output: 2

##Comparison Operators

x = 10
y = 20

# Equal to
print("Equal to:", x == y)  # Output: False

# Not equal to
print("Not equal to:", x != y)  # Output: True

# Greater than
print("Greater than:", x > y)  # Output: False

# Less than
print("Less than:", x < y)  # Output: True

# Greater than or equal to
print("Greater than or equal to:", x >= y)  # Output: False

# Less than or equal to
print("Less than or equal to:", x <= y)  # Output: True

##Logical Operators

a = True
b = False

# AND
print("AND:", a and b)  # Output: False

# OR
print("OR:", a or b)  # Output: True

# NOT
print("NOT:", not a)  # Output: False

#Bitwise Operators

a = 5  # Binary: 0101
b = 3  # Binary: 0011

# AND
print("AND:", a & b)  # Output: 1 (Binary: 0001)

# OR
print("OR:", a | b)  # Output: 7 (Binary: 0111)

# XOR
print("XOR:", a ^ b)  # Output: 6 (Binary: 0110)

# NOT
print("NOT:", ~a)  # Output: -6 (Binary: ...11111010)

# LEFT SHIFT
print("LEFT SHIFT:", a << 1)  # Output: 10 (Binary: 1010)

# RIGHT SHIFT
print("RIGHT SHIFT:", a >> 1)  # Output: 2 (Binary: 0010)

#Assignment Operators
x = 10

# Assignment

x = 5
print("Assignment:", x)  # Output: 5

# Add and assign
x += 5
print("Add and assign:", x)  # Output: 10

# Subtract and assign
x -= 3
print("Subtract and assign:", x)  # Output: 7

# Multiply and assign
x *= 2
print("Multiply and assign:", x)  # Output: 14

# Divide and assign
x /= 7
print("Divide and assign:", x)  # Output: 2.0

# Modulus and assign
x %= 2
print("Modulus and assign:", x)  # Output: 0.0

# Exponentiate and assign
x **= 3
print("Exponentiate and assign:", x)  # Output: 0.0

# Floor divide and assign
x //= 1
print("Floor divide and assign:", x)  # Output: 0.0

##Identity Operators

a = [1, 2, 3]
b = [1, 2, 3]
c = a

# is
print("a is b:", a is b)  # Output: False

# is not
print("a is not b:", a is not b)  # Output: True

# is
print("a is c:", a is c)  # Output: True

# is not
print("a is not c:", a is not c)  # Output: False

#Membership Operators

list_ = [1, 2, 3, 4, 5]

# in
print("3 in list_:", 3 in list_)  # Output: True

# not in
print("6 not in list_:", 6 not in list_)  # Output: True



```

```python
#Q5. solution of Q5
#Example of Implicit Type Casting
x = 5       # int
y = 2.0     # float

result = x + y  # x is implicitly converted to float
print(result)   
print(type(result)) 

#Example of Explicit  Type Casting
#1.Converting String to Integer:
s = "123"
x = int(s)
print(x)        
print(type(x))  

#2.Converting Float to Integer:
f = 3.14
x = int(f)
print(x)        # Outputs: 3
print(type(x))

#3.Converting Integer to String:
n = 100
s = str(n)
print(s)        # Outputs: "100"
print(type(s))  # Outputs: <class 'str'>

#4Converting List to Tuple:
lst = [1, 2, 3]
t = tuple(lst)
print(t)        # Outputs: (1, 2, 3)
print(type(t))  # Outputs: <class 'tuple'>

#5.Converting Tuple to List:
t = (1, 2, 3)
lst = list(t)
print(lst)        # Outputs: [1, 2, 3]
print(type(lst))  # Outputs: <class 'list'>

#6.Converting List to Set:
lst = [1, 2, 2, 3]
s = set(lst)
print(s)        # Outputs: {1, 2, 3}
print(type(s))  # Outputs: <class 'set'>
```

```python
#Q.6 Solution of Q6. 
#Example
# 'if' Statement:
x = 10

if x > 5:
    print("x is greater than 5")
    
# 'if-else' Statement:
x = 3

if x > 5:
    print("x is greater than 5")
else:
    print("x is 5 or less")
    
# 'if-elif-else' Statement:
x = 7

if x > 10:
    print("x is greater than 10")
elif x > 5:
    print("x is greater than 5 but 10 or less")
else:
    print("x is 5 or less")
    
# 'Nested if' Statements:
x = 15
y = 10

if x > 10:
    if y > 5:
        print("x is greater than 10 and y is greater than 5")
    else:
        print("x is greater than 10 but y is 5 or less")
else:
    print("x is 10 or less")
    
#Ternary Conditional Operator:
x = 8
result = "x is even" if x % 2 == 0 else "x is odd"
print(result)

```

```python
#Q7. Solution of Q7.
#for loop example 
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
    
# Iterating over a list:
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
    
#Using range() function:    
for i in range(5):
    print(i)
    
#Iterating over a dictionary:   
student_grades = {"Yash": 90, "Suraj": 85, "Amar": 88}
for student, grade in student_grades.items():
    print(f"{student}: {grade}")
    
#Basic 'while loop':
count = 0
while count < 5:
    print(count)
    count += 1
    
#Stopping condition:
user_input = ""
while user_input != "exit":
    user_input = input("Type 'exit' to stop: ")
    print("You typed:", user_input)
    
```
